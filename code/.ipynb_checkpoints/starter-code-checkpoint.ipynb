## Background

The SAT and ACT are standardized tests that many colleges and universities in the United States require for their admissions process. This score is used along with other materials such as grade point average (GPA) and essay responses to determine whether or not a potential student will be accepted to the university.

The SAT has two sections of the test: Evidence-Based Reading and Writing and Math ([*source*](https://www.princetonreview.com/college/sat-sections)). The ACT has 4 sections: English, Mathematics, Reading, and Science, with an additional optional writing section ([*source*](https://www.act.org/content/act/en/products-and-services/the-act/scores/understanding-your-scores.html)). They have different score ranges, which you can read more about on their websites or additional outside sources (a quick Google search will help you understand the scores for each test):
* [SAT](https://collegereadiness.collegeboard.org/sat)
* [ACT](https://www.act.org/content/act/en.html)

Standardized tests have long been a controversial topic for students, administrators, and legislators. Since the 1940's, an increasing number of colleges have been using scores from sudents' performances on tests like the SAT and the ACT as a measure for college readiness and aptitude ([*source*](https://www.minotdailynews.com/news/local-news/2017/04/a-brief-history-of-the-sat-and-act/)). Supporters of these tests argue that these scores can be used as an objective measure to determine college admittance. Opponents of these tests claim that these tests are not accurate measures of students potential or ability and serve as an inequitable barrier to entry. Lately, more and more schools are opting to drop the SAT/ACT requirement for their Fall 2021 applications ([*read more about this here*](https://www.cnn.com/2020/04/14/us/coronavirus-colleges-sat-act-test-trnd/index.html)).

ACT was introduced in late 1959 by ACT, Inc. and has since gained steady traction, with Colorado and Illinois being one of the first to implement it as a state-wide test in 2001 and numerous other states following suit ([source](https://www.act.org/content/act/en/about-act.html)). The ACT was marketed as a test to replace and standardize the graduation requirements for high school juniors with the intent to give every student access to a college admissions test and motivate more students to consider college upon graduation ([source](https://blog.prepscholar.com/which-states-require-the-act-full-list-and-advice)). Currently,13 states require high school juniors to sit the test while 8 others offer it as an option, at no extra cost to the student ([source](https://blog.prepscholar.com/which-states-require-the-act-full-list-and-advice)).

SAT was developed and introduced in 1926 by the College Board with major revisions being made to the test in 2016 to better prepare students for the reading and writing standards expected at college level ([source](https://sat.ivyglobal.com/new-vs-old)). Over time, the test has gained more popularity with tests becoming administered at no cost to students and better test-prep resources becoming readily available ([source](https://www.wsj.com/articles/sat-scores-fall-as-more-students-take-the-test-11569297660)). 20 states have made it a graduation requirement for high school juniors while 7 states offer it but do not require it ([source](https://blog.prepscholar.com/which-states-require-the-sat)).

1. Manually calculate mean:

    Write a function that takes in values and returns the mean of the values. Create a list of numbers that you test on your function to check to make sure your function works!

```python
def find_mean(numbers):
    return sum(numbers)/len(numbers)
num1 = [2,4,8,16,17]

find_mean(num1)
```

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
def stdev(numbers):
    mean = find_mean(numbers)
    deviations = [(x-mean)**2 for x in numbers]
    variance = sum(deviations)/len(numbers)
    return variance**0.5

stdev(num1)
```

#### 3a) EDA: ACT OBSERVATIONS

* Participation rate

Alabama, Kentucky, Wisconsin, Utah and Tennessee maintained a 100% participation rate from 2017-2018. These states all require the ACT as a graduation requirement though Tennessee also offers the SAT as an alternative. 

North and South Dakota as well as Hawaii show consistency in ranking high in participation rates across 2017-2018 for the ACT. North Dakota and Hawaii require the ACT. However, South Dakota as well as Kansas do not. Nonetheless, it seems to be the traditionally [preferred]( https://www.studypoint.com/ed/sat-and-act-test/) college admissions test in the Midwest. Interestingly, Illinois, which traditionally preferred the ACT, signed with the college board to administer the [SAT in 2016](https://chicago.chalkbeat.org/2018/7/27/21105418/illinois-has-embraced-the-sat-and-the-act-is-mad-about-it). While it still has a 93% participation rate in 2017, it drops to 43% in 2018 showing that the state is moving away from the ACT and embracing the SAT. However, there has been debate of biasness shown by the state board in choosing the college boards contract over ACT inc.s and given that the contract lasts for 3 years, there is a possibility to explore how ACT inc. can once again bid to establish its stronghold on Illinois.

The same 5 states- Maine, Rhode Island, New Hampshire, Delaware and Pennsylvania, showed consistency low participation rates in the ACT from 2017-2018. Apart from Rhode Island and Pennsylvania, all states require the SAT hence explaining the low ACT participation rates. The former two states show a consistency high levels of participation in the SAT across 2017-2018 as is characteristic of East Coast. 

* What does the year to year participation change tell

The 16% increase in participation from 2017-2018 for Nebraska is a direct result of the state [requiring the ACT]( https://www.education.ne.gov/assessment/act/) to be administered from the beginning of spring 2017. Ohio offers both the ACT and SAT but a strong preference for the ACT can be observed given its 25% increase in participation from 2017-2018.

Colorado shows a whopping 70% decrease in participation (previously 100% in 2017), Illinois  50% and Michigan 7% - all these scenarios can be attributed to the [college board winning contracts]( https://thecriticalreader.com/for-what-its-worth/) over the ACT inc. with the respective state boards in 2015-2016. The SAT gaining traction seems closely linked to the revision of the SAT format in 2016 which aimed to be more aligned with the Core Curriculum, making it more relevant to college-bound high schoolers. 

Alaska is interesting in that their contract to administer the SAT or ACT [expired in 2016]( https://www.adn.com/alaska-news/education/2016/06/30/students-no-longer-need-national-tests-to-graduate/) so neither test was required from the following year onwards. This explains the 32% decrease in participation in the ACT. The likely reason may be that students now have to pay out of pocket for the tests themselves which can be a deterrent. The state board could potentially be persuaded to sign with ACT inc. if they can be convinced that their investment in education does yield positive outcomes in the form of better student participation/ student motivation levels to do the test and higher number of students wanting to pursue college studies. 

* Observing the relationship between composite scores and relative participation rate

States with close to 100%  participation in both 2017 and 2018, seem to have some of the lowest mean composite scores while states with less than 35% participation seem to yield the highest mean composite scores. 

There is a simple elaboration for the results  [here]( https://www.compassprep.com/great-to-good-the-diluted-value-of-high-test-scores/). To summarise, when more students are state-mandated to take a test, they take a test for the sake of it as opposed to those who opt to take the test. This means that though there might be 100% participation, they are not all performing to the best of their ability. States where participation was lower but mean composite scores were higher probably had a smaller group of students with a focused intent on getting into college a.k.a high achievers. Additionally, when more students take a test, the distribution of scores tends to become more diluted which means it may not be a good or true representation of the states performance when compared to say the median which might be a better indicator.

## 2) Histograms

```python
#Setting up the subplot
#Reference :https://towardsdatascience.com/master-the-art-of-subplots-in-python-45f7884f3d2e
def subplot_histograms(dataframe, list_of_columns, list_of_titles, list_of_xlabels,
                      list_of_ylabels, figsize, bin_cnt, xlim, ylim):
   
    nrows = int(np.ceil(len(list_of_columns)/2))
    
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize = figsize)
    
    ax = ax.ravel()
    
    for i, column in enumerate(list_of_columns):
        ax[i].hist(dataframe[column], bins = bin_cnt) 
        
        ax[i].tick_params(axis="x", labelsize=14)
        ax[i].tick_params(axis="y", labelsize=14)
        ax[i].set_title(list_of_titles[i], fontsize = 20)
        ax[i].set_xlabel(list_of_xlabels[i], fontsize = 16)
        ax[i].set_ylabel(list_of_ylabels[i], fontsize = 16)
        ax[i].set_xlim(xlim[i])
        ax[i].set_ylim(ylim[i])
        ax[i].grid(False)
```

```python
# Histograms of SAT and ACT participation (2017 & 2018)
subplot_histograms(act_sat_final,
                   ["act_participation_2017","act_participation_2018",
                    "sat_participation_2017","sat_participation_2018"],
                   ["ACT participation 2017", "ACT participation 2018",
                    "SAT participation 2017", "SAT participation 2018"],
                   ["Participation rate", "Participation rate",
                    "Participation rate", "Participation rate"],
                   ["No. of States", "No. of States", "No. of States","No. of States"],
                  (20,18), 
                  10,
                  [[0,1],[0,1],[0,1],[0,1]],
                  [[0,25],[0,25],[0,25],[0,25]])
```

```python
# Histograms of SAT and ACT scores (2017 & 2018)
subplot_histograms(act_sat_final,
                   ["act_composite_2017","act_composite_2018",
                    "sat_total_2017","sat_total_2018"],
                   ["ACT composite scores 2017", "ACT composite scores 2018",
                    "SAT total scores 2017", "SAT total scores 2018"],
                   ["Composite scores", "Composite scores",
                    "Total scores", "Total scores"],
                   ["No. of States", "No. of States", "No. of States","No. of States"],
                  (20,18), 
                  10,
                  [[15,30],[15,30],[900,1400],[900,1400]],
                  [[0,15],[0,15],[0,15],[0,15]])
```

### Histogram OBSERVATIONS

* Participation rates

Observing the participation rates in the ACT and SAT over 2017-2018, we can see that both tests express bimodal tendencies. This means states have either high (100%) or low (<30%) participation rates with few data points falling in between. 

* ACT mean composite scores

More states seem to fall between the 18-24 score range and 24-26 score range with very few falling in between these ranges. There seem to be two peaks forming as a result.

More states seem to fall between the 950-1150 score range and 1200-1300 score range with very few falling in between there ranges. There seem to be two peaks forming as a result.
